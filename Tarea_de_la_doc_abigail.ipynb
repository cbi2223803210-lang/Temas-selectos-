# ** Problema 6 **

Uso de la condicionales para formula de reynolds



```python
p=1000 # kg*cm3
v=1.2 # m s-1
D=0.05 # m
mu=0.001 #Pa*s
Re=p*v*D/mu
print(f"El valor del numero adimensional",Re)

if 0<Re<2300:
     print("El flujo es laminar")
elif 2300<=Re<=4000:
     print("El flujo es transiente")
elif Re>4000:
     print("El flujo es transitorio")

else:
     print("Revisa datos")
```
